```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.set()
df=pd.read_csv('OpidsVA.csv')
df1=pd.read_csv('presidents.csv')
df2=pd.read_csv('TOTALNSA.csv')
```

```python
plt.scatter(df['MedianHouseholdIncome'],df['FPOO-Rate'])
plt.ylabel('FPOO-Rate (Overdoses)')
plt.xlabel('Median-Household Income')
plt.title('Median-Household Income Vs. Opioid Overdoses');
```

```python
# there is a negative weak relationship between median household income and opioid overdoses 
# from the graph we can summize that less affluent households do see a rise in overdoses however 
#there exists a subtle relationship between the two variables in question. 
```

```python
plt.scatter(df['FFHO-Rate'],df['FPOO-Rate'])
plt.xlabel('FFHO-Rate')
plt.ylabel('FPOO-Rate')
plt.title('Heroin vs. Opioid Overdose Rates');
```

```python
# There exists a positive weak relationship between opioid overdoses and heroin overdoses.
```

```python
x=df1.groupby('State')['State'].count().sort_values()
x.plot(kind='barh',grid=False)
plt.title('U.S. Presidents Birthplace by State')
plt.xlabel('Frequency');
```

```python
##The states associated with the greatest number of presidents in the United States in terms
## of birthplaces are Virginia,Ohio,New York and Massachusetts.
```

```python
df2['DATE']=pd.to_datetime(df2['DATE'])
plt.plot(df2['DATE'].dt.year,df2['TOTALNSA'],color='black')
plt.title('Vehicle Sales (1976-2018)')
plt.xlabel('Years')
plt.ylabel('Total Sales');
```

```python
# There is a noticeable upward trend in vehicle sales since january 1st,1976 until September 1st,2018.
# There is also an element of seasonality to the data showing that vehicle sales tend to increase 
# during the colder months,namely December-February, in comparison to warmer months.
```
